```python
import matplotlib.pyplot as plt
import pandas as pd
path = r"F:\AI\ML\DATA\smoke_detection_iot.csv"
data = pd.read_csv(path)
count_yesfire=0
count_Nofire=0
for column in data["Fire_Alarm"]:
    if column==1:
        count_yesfire+=1
    else:
        count_Nofire+=1
sum=count_yesfire+count_Nofire
p_count_yesfire=(count_yesfire*100)/sum
p_count_Nofire=(count_Nofire*100)/sum
print(p_count_yesfire)
print(p_count_Nofire)
colors = ['#FDEE58', '#4B77BE']
labels = ['Yes Fire', 'No Fire']
sizes = [p_count_yesfire, p_count_Nofire]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
ax1.set_title('Fire Alarm Data')
plt.legend(labels, loc='lower left')
plt.show()
```
